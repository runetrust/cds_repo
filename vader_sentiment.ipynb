```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/work/CulturalDataScience/AB_NYC_2019.csv")
```

```python
sentences = data["name"]
```

```python
analyzer = SentimentIntensityAnalyzer()

positive = pd.DataFrame(columns=["title", "vs"])
neutral = pd.DataFrame(columns=["title", "vs"])
negative = pd.DataFrame(columns=["title", "vs"])

for i, sentence in enumerate(sentences):
    if not isinstance(sentence, str):
        print("Non-string detected at index", i, ":", sentence, type(sentence))
        continue
    vs = analyzer.polarity_scores(sentence)
    print("{:-<65} {}".format(sentence, str(vs)))
    if vs["compound"] > 0.5:
        positive.loc[len(positive)] = [sentence, vs]
    elif -0.5 <= vs["compound"] < 0.5:
        neutral.loc[len(neutral)] = [sentence, vs]
    else:
        negative.loc[len(negative)] = [sentence, vs]
    
```

```python
rows = []   # temporary storage for all results

analyzer = SentimentIntensityAnalyzer()

for i, sentence in enumerate(sentences):
    if not isinstance(sentence, str):
        print("Non-string detected at index", i, ":", sentence, type(sentence))
        continue
    vs = analyzer.polarity_scores(sentence)

    # classify
    if vs["compound"] > 0.5:
        sentiment = "positive"
    elif -0.5 <= vs["compound"] < 0.5:
        sentiment = "neutral"
    else:
        sentiment = "negative"

    # store row
    rows.append({
        "name": sentence,
        "vs_neg": vs["neg"],
        "vs_pos": vs["pos"],
        "vs_neu": vs["neu"],
        "vs_compound": vs["compound"],
        "sentiment": sentiment
    })

# build dataframe once
df = pd.DataFrame(rows)


#final_df = data.merge(df, on="name", how="right", suffixes=("_data", "_sent"))
```
